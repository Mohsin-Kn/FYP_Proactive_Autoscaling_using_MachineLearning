```python
df = pd.read_csv('30_day_httpRequests.csv',nrows=360)

# Convert the timestamp column to datetime type
df['timestamp'] = pd.to_datetime(df['timestamp'])

# Set timestamp as the index
df.set_index('timestamp', inplace=True)

# Perform time series decomposition to extract trend, seasonality, and residuals
# Adjust the period parameter based on the df frequency (e.g., 24 for hourly df with daily seasonality)
decomposition = seasonal_decompose(df['http_requests'], model='additive', period=24)

# Plot the decomposition
plt.figure(figsize=(12, 8))
decomposition.plot()
plt.show()
```

```python
# Load your dataset
df = pd.read_csv('30_day_httpRequests.csv')
df['timestamp'] = pd.to_datetime(df['timestamp'])
df.set_index('timestamp', inplace=True)

# Plot ACF
plt.figure(figsize=(10, 6))
plot_acf(df['http_requests'], lags=48, ax=plt.gca())  # Adjust 'lags' as needed
plt.title('Autocorrelation Function (ACF)')
plt.show()
```
